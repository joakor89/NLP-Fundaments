### StopWords Fraction on Corpus

```python
def stopwords_percentage(text):
  '''
  Lexical Resource Use (stopwords) to Corpus Filtering
  '''
  stopwd = stopwords.words('english')
  content = [w  for w in text if w.lower() not in stopwd]
  return 1 - len(content)/len(text)
```

```python
# Synset: Single Word Synonyms Group
ss = wn.synsets('carro', lang='spa')
ss
```

```python
# Synset Exploratory Approach
for syn in ss:
  print(syn.name(), ': ', syn.definition())
  for name in syn.lemma_names():
    print(' * ', name)
```

```python
# Visualization References
# * http://www.randomhacks.net/2009/12/29/visualizing-wordnet-relationships-as-graphs/
# * http://dlacombejr.github.io/programming/2015/09/28/visualizing-cifar-10-categories-with-wordnet-and-networkx.html

import networkx as nx
import matplotlib.pyplot as plt

def closure_graph(synset, fn):
    seen = set()
    graph = nx.DiGraph()
    labels = {}

    def recurse(s):
        if not s in seen:
            seen.add(s)
            labels[s.name] = s.name().split('.')[0]
            graph.add_node(s.name)
            for s1 in fn(s):
                graph.add_node(s1.name)
                graph.add_edge(s.name, s1.name)
                recurse(s1)

    recurse(synset)
    return graph, labels

def draw_text_graph(G, labels):
    plt.figure(figsize=(18,12))
    pos = nx.planar_layout(G, scale=18)
    nx.draw_networkx_nodes(G, pos, node_color="yellowgreen", linewidths=0, node_size=300)
    nx.draw_networkx_labels(G, pos, font_size=14, labels=labels)
    nx.draw_networkx_edges(G, pos)
    plt.xticks([])
    plt.yticks([])
```

### **Hyponyms:** Concepts that are more specific than the root word from which they are derived.

```python
print(ss[0].name())
G, labels = closure_graph(ss[0], fn = lambda s: s.hyponyms())
draw_text_graph(G, labels)
```

### **Hypernyms**: More General Concepts

```python
print(ss[0].name())
G, labels = closure_graph(ss[0], fn = lambda s: s.hypernyms())
draw_text_graph(G, labels)
```

### Semantic Similarity

```python
def show_syns(word):
  ss = wn.synsets(word, lang='spa')
  for syn in ss:
    print(syn.name(), ': ', syn.definition())
    for name in syn.lemma_names():
      print(' * ', name)
  return ss
```

```python
ss = show_syns('perro')
```

```python
ss2 = show_syns('gato')
```

```python
ss3 = show_syns('animal')
```

```python
# Setting Up Variables 

# Dog
perro = ss[0]
# Cat
gato = ss2[0]
# Animal
animal = ss3[0]
```

```python
# Similarity between 'animal' & 'perro' dog - Distance Vertex
animal.path_similarity(perro)
```

```python
# Similarity between 'animal' & 'gato' cat - Distance Vertex
animal.path_similarity(gato)
```
