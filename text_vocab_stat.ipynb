# Measure of lexical richness in a text:

---
*   $$ R_l = \frac{\text{total unique words}}{\text{total words}} = \frac{\text{vocabulary length}}{\text{text length}}$$

```python
# Let's build a vocabulary (Identifying book's unique words)
# https://docs.python.org/2/library/sets.html
vocabulary = sorted(set(text1))
vocabulary[1000:1050]
```

```python
# Lexical Richness Measurement
lr = len(set(text1))/len(text1)
print(lr)
```

```python
# Lexical Measurement Function
def lexical_rich(texto):
  return len(set(texto))/len(texto)

def word_percentage(words, texto):
  return 100*texto.count(words)/len(texto)

lexical_rich(text1)
```

```python
# Checking Word % within a book
word_percentage('monster', text1)
```

```python
# Counting Repetition
text1.count('monster')
```

# Language Statistics

---
*   The simplest statistical calculations that can be performed on a text or a corpus are those related to the frequency of occurrence of words.
*   We can build a dictionary in Python where the keys are the words and the values are the frequencies of occurrences of those words.
*   For isntance: `dic = {'monster': 49 ,  'boat': 54,  ...}`

```python
# NOT RECCOMENDED METHOD tailored to Massive Datasetsd 
dic = {}
for word in set(text1):
  #dic[word] = word_percentage(word, text1)
  dic[word] = text1.count(word)
dic
```

# Fine-Filter Content Distribution

---
*   As we saw in the previous section, the most frequent tokens in a text are not necessarily the words that give us the most information about its content.
*   Therefore, it is better to filter and build frequency distributions that do not consider punctuation marks or special characters.

```python
# Minimun Specific Length
long_words = [word for word in text1 if len(word)>5]
# Filtered Vocabulary
filtered_vocabulary = sorted(set(long_words))
filtered_vocabulary[:10]
```

We can now build a list of tuples, count's based already pre-computed from the `FreqDist` function, like so:

*   `fdist_filtered = [('account', 2), ('additional', 32), ('advancing', 5), ...]`
---
Then convert the created object on **Numpy**'s `np.array` format in order to sort by most frequent words.

*   `fdist_filtered = np.array(fdist_filtered, *args)`
*   `np.sort(fdist_filtered, *args) = [('account', 2), ('advancing', 5), ('additional', 32), ...]`



```python
# Setting Iterative Perambulate Proceduce
interesting_words = [(word, fdist[word]) for word in set(text1) if len(word)>5 and fdist[word]>10]
# List Elements Interpretation
dtypes = [('word', 'S10'), ('frequency', int)]
# Array's Conversion
interesting_words = np.array(interesting_words, dtype=dtypes)
# Sorting & Fequency
ordered_words = np.sort(interesting_words, order = 'frequency')
# Top Most Frequent Words
ordered_words[-20:] 
```

```python
# Matplotlib Customized Visualization
# https://matplotlib.org/3.1.1/gallery/ticks_and_spines/custom_ticker1.html#sphx-glr-gallery-ticks-and-spines-custom-ticker1-py

top_words = 20
# X Axe
x = np.arange(len(ordered_words[-top_words:]))
# Y axe
y = [freq[1] for freq in ordered_words[-top_words:]]

# FigSize Standard
plt.figure(figsize=(10,5))
plt.plot(x, y, color='yellowgreen')
plt.xticks(x, [str(freq[0]) for freq in ordered_words[-top_words:]], rotation = 'vertical')
plt.tick_params(labelsize = 'xx-large')
# Grid Standard
plt.grid(True)
plt.show()
```
